## Backpropagation in code

```python
import numpy as np
```

```python
# Create a dataset 

# X = (hours of sleeping, hours of studying), y = test score of the studnet 

X = np.array(([2, 9], # Inputs
              [1, 5], 
              [3, 6]) 
             ,dtype=float  ) 


y = np.array( # Output
            ([92], 
             [86], 
             [89])
             ,dtype=float) 


# Scale units 
X = X/np.amax(X, axis=0) # maximum of x array. axis = 0 NumPy finds the maximum value for each column
y = y/100 # maximum test score is 100
```

```python
class NeuralNetwork(object):

    def __init__(self):
        
        # parameters 

        self.inputSize = 2
        self.outputSize = 1
        self.hiddenSize = 3

        # Because we have more than one input and one hidder neuron, we have alot of connections, so we create a matrix 
        # of weights --> Number of weights = number of connections between neurons.

        # Weights here it will be a mtrix not just a single number this is the net between the inputs layer and the hidden layer (3×2) matrix
        # Weight matrix shape=(number of neurons in previous layer,number of neurons in next layer)
        self.W1 = np.random.randn(self.inputSize, self.hiddenSize) # Because in backpropagation first we assign the weights randomly (3×2) 
        self.W2 = np.random.randn(self.hiddenSize, self.outputSize) # (3×1) from hidden layer to output layer 

    def feedForward(self, X):

        # forward propagation through the network 

        
        # The dot product between the input and the weight matrix (step 1)
        self.z = np.dot(X, self.W1) # (3×2)

        # The activation function (step 2)
        self.z2 = self.sigmoid(self.z)

        # The dot product between the hidden layer and the second weight matrix  
        self.z3 = np.dot(self.z2, self.W2) # (3×1)


        output = self.sigmoid(self.z3) # Without sigmoid: output = 3.2, This does not make sense as a class probability.
        # With sigmoid: sigmoid(3.2) = 0.96 --> Now you can interpret it as: 96%
        return output


    # Takes a number and turn it into a probability (0-1)   
    def sigmoid(self, num, derivative = False): # derivative is a flag (True/False) that decides whether you want:
                                                # the sigmoid value
                                                # or its derivative
        if (derivative == True):
            return num * (1 - num) # derivative of the sigmoid function. How much can this neuron output change if we slightly change its input?
        else:
            return 1/(1 + np.exp(-num)) # The sigmoid function 


    # We are trying to calculate the loss function (the sum of squares) and minimize it 
    def backward(self, X, y, output):
        # Backward propagate through the network 

        self.nn_output_error = y - output # y --> is the true value and output is the prediction 

        self.output_delta = self.nn_output_error * self.sigmoid(output, derivative = True)

        self.z2_error = self.output_delta.dot(self.W2.T) # z2 error: how much the hidden layer weights contribute to output error 
        self.z2_delta = self.z2_error * self.sigmoid(self.z2, derivative = True) # # applying derivative of sigmoid to z2 error \
    
        self.W1 += X.T.dot(self.z2_delta) # adjusting first set (input --> hidden) weights 

        self.W2 += self.z2.T.dot(self.output_delta)

    
    def train(self, X, y):

        output = self.feedForward(X)
        self.backward(X, y, output)

NN = NeuralNetwork()

for i in range(1000): # Trains the NN 1000 times 
    NN.train(X, y)

print("Input: " + str(X))
print()
print("Actual Output: " + str(y))
print()
print("Loss: " + str(np.mean(np.square(y - NN.feedForward(X)))))
print()
print("Predicted Output: " + str(NN.feedForward(X)))
```
